```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
df = pd.read_csv('./data/kc_house_data.csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(x=df['sqft_living'],y=df['price'])
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(x=df['grade'],y=df['price'])
plt.xlabel('grade')
plt.ylabel('price')
plt.show()
```

```python
# Plotting the correlation matrix
# values close to 1 indicate strong positive correlation,close to -1 indicate strong negative correlation
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Suppose you just want to see correlation with the price variable and not the full matrix
df.corr()['price'].sort_values(ascending = False)
```

```python
# Performs linear regression based on the selected features
def linear_regression(features):
    new_data = df[features]
    X = new_data.values
    y = df.price.values
    # Split the data into training and testing sets
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 4)
    # Model training
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    predictions = lr.predict(X_test)
    print(f'Root Mean Squared Error: {np.sqrt(metrics.mean_squared_error(predictions,y_test))}')
```

```python
# Importance of features used
linear_regression(['sqft_living'])
```

```python
linear_regression(['sqft_living','bedrooms','bathrooms'])
```

```python
# Relationship between the waterfront feature and price of the house
sns.boxplot(df['waterfront'],df['price'])
```

```python
linear_regression(['sqft_living','bedrooms','bathrooms','waterfront'])
```

```python
# Let's add grade to the list of features and observe the result
# The reason for adding grade is that it has very high correlation with price
linear_regression(['sqft_living','bedrooms','bathrooms','waterfront','grade'])
```

```python
sns.boxplot(df['condition'],df['price'])
```

```python
# Add the condition feature
linear_regression(['sqft_living','bedrooms','bathrooms','waterfront','grade','condition'])
# Again a reduction in error
```

```python
# Now let's build the final model based everything observed so far
linear_regression(['sqft_living','bedrooms','bathrooms','waterfront','grade','condition','lat','sqft_living15'
                  ,'lat','view'
                  ])
```
